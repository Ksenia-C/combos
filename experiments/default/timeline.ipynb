```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
import seaborn as sns
import scipy as scp
import pylab as pl
from matplotlib import collections as mc
```

```python
gl_colors = {
    "unavailable:": 'black',
     "idle:": 'red', 
     "busy:Project1": 'green',
     "busy:Project2": 'blue'}
    
```

```python
data_execs = pd.read_csv(f"execute_stat_4.csv")
```

```python
def draw_for_client(client_name):
    lines = []
    colors = []
    total_pr = 5
    for ind_expr in range(total_pr) :
        data_execs = pd.read_csv(f"execute_stat_{ind_expr}.csv")
        data_execs = data_execs[data_execs['client'] == client_name]
        
        for type in data_execs.type.unique():
            values_of_type = data_execs[data_execs.type == type]
            starts = values_of_type.start.values
            ends = values_of_type.end.values
            
            for i in range(values_of_type.shape[0]):
                lines.append([(starts[i], total_pr - ind_expr), (ends[i], total_pr - ind_expr)])
                colors.append(gl_colors[type])



    lc = mc.LineCollection(lines, colors=np.array(colors), linewidths=18)
    fig, ax = pl.subplots(figsize=(18,4))
    ax.add_collection(lc)
    ax.autoscale()
    ax.margins(0.1)
    # ax.set_xlim([29400, 29600])
    plt.show()
```

```python
draw_for_client("c10")
```
