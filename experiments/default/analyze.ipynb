```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
import seaborn as sns
import scipy as scp
```

```python
data = pd.read_csv("result.csv")
x = data['proj1_nvalid_result'].values
y = data['proj2_nvalid_result'].values
types = data['label'].values

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(x, y)

ax.set_xlabel('nvalidr for smaller tasks', fontsize=14)
ax.set_ylabel('nvalidr for larger tasks', fontsize=14)
ax.set_title('Proportion of results based on priorities', fontsize=18)

for i, txt in enumerate(types):
    ax.annotate(txt, (x[i], y[i]), xytext=(10,10), textcoords='offset points')
    plt.scatter(x, y, marker='x', color='red')
```
